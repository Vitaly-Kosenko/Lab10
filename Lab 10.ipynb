<h2>1). Створити додаток, що вводить 2 списки a і b. Необхідно повернути список,
що складається з елементів, що є загальними для обох списків.</h2>

```python
a = [1, 2, 3, 4, 5, 6, 7]
b = [i for i in a if i % 2 == 0]
print('a = {}\nb = {}'.format(a, b))
```

<h2>4). Створити додаток, що вводить цілі числа n і m підрахувати не обов’язково повинно бути однорозрядним
числом. Наприклад n = 10, nn = 1010, nnn = 101010 і т.д.</h2>

```python
a = int(input("Перший номер: ")) # ввод первого числа
b = int(input("Друге число: ")) # ввод второго числа
z = [] # пустой список
while a < b:
    if a < b:
        a = a + 2
        z.append(a)
    elif a == b:
        break
print(z)
```

<h2>5). Створити додаток для виведення n рядків трикутника Паскаля. n вводиться
з клавіатури.</h2>

```python
count = 0 
def Pascal(n):
    global count
    count += 1 

    pasc_list = [] 
    i = 0 
    j = i+1
    dictionary = {0:[1],1:[1,1]}

    if n in dictionary:
        return dictionary[n]

    else:
        pasc_list.append(1)
        p = Pascal(n-1)  
        while j < len(p):
            pasc_list.append(p[i] + p[j])
            i += 1 
            j = i+1 

        pasc_list.append(1)
        dictionary[n] = pasc_list

    return pasc_list

a = Pascal(7)
print(a)
print(count)
```
